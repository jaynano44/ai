## 2.2 LangChain을 선택

```python
from langchain_openai import ChatOpenAI
# .env에 OPEN_API_KEY 등록
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.5
)
ai_message = llm.invoke("2020년 월드 시리즈는 누가 우승했어요?")
```

```python
ai_message
```

```python
ai_message.content
```

```python
ai_message.usage_metadata
```

```python
from langchain_core.messages import SystemMessage,HumanMessage
messages = [
    SystemMessage(content="너는 친절하게 대답해 주는 비서야"),
    HumanMessage(content="2020년 월드 시리즈는 누가 우승했어?")
]
ai_message = llm.invoke(messages)
print(ai_message.content)
```
